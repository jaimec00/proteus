```python
from proteus.data.data_loader import DataHolder, DataHolderCfg
from proteus.data.construct_registry import ConstructRegistry, ConstructFunctionNames
import logging
logging.basicConfig(level=logging.INFO)

ConstructRegistry.set_construct_function(ConstructFunctionNames.PROTEUS)

cfg = DataHolderCfg(
    s3_bucket="s3://proteus-data-bucket",
    cluster_split_cols=["foldseek_70", "mmseqs_30"],
    cluster_col="foldseek_70",
    num_workers=0,
)

data_holder = DataHolder(cfg)
```

```python
data = iter(data_holder.train)
```
