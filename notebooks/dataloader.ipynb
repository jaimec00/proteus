# dataloder stuff

### for debugging the dataloader

```python
from proteus.data.data_loader import DataHolder, DataHolderCfg
from proteus.data.data_constants import IndexCol

cfg = DataHolderCfg(
    s3_bucket="s3://proteus-data-bucket",
)

data_holder = DataHolder(cfg)

for index, name in [
    (data_holder.raw_index, "raw_index"), 
    (data_holder.index, "index"), 
    (data_holder.train_index, "train_index"), 
    (data_holder.val_index, "val_index"), 
    (data_holder.test_index, "test_index")
]:
    cols = index.columns
    num_rows = len(index)
    clusters = index[IndexCol.CLUSTER_ID].unique()
    num_clusters = len(clusters)

    print(
        f"name: {name}",
        f"columns: {cols}",
        f"num_rows: {num_rows}",
        f"num_clusters: {num_clusters}",
        "-"*20,
        sep="\n"
    )

print("\n-- SPLIT STATISTICS --\n")

tot_train, tot_clusters_train = len(data_holder.train_index), len(data_holder.train_index[IndexCol.CLUSTER_ID].unique()) 
tot_val, tot_clusters_val = len(data_holder.val_index), len(data_holder.val_index[IndexCol.CLUSTER_ID].unique()) 
tot_test, tot_clusters_test = len(data_holder.test_index), len(data_holder.test_index[IndexCol.CLUSTER_ID].unique()) 
tot = tot_train + tot_val + tot_test
tot_cluster = tot_clusters_train + tot_clusters_val + tot_clusters_test

for name, samples, clusters, trgt_pct in [
    ("train", tot_train, tot_clusters_train, cfg.train_val_test_split[0]),
    ("val", tot_val, tot_clusters_val, cfg.train_val_test_split[1]),
    ("test", tot_test, tot_clusters_test, cfg.train_val_test_split[2]),
]:
    print(
        f"name: {name}",
        f"trgt_pct: {trgt_pct}",
        f"pct_sample: {samples / tot}",
        f"pct_clusters: {clusters / tot_cluster}",
        "-"*20,
        sep="\n"
    )

```
